```python
df = df.sort_values(["player_id", "round"])
```

```python
df.head()
```

##Average points from last 3


```python
df["avg_points_last3"] = (
    df.groupby("player_id")["total_points"]
    .shift(1)
    .rolling(3)
    .mean()
)
```

```python
df["avg_minutes_last3"] = (
    df.groupby("player_id")["minutes"]
    .shift(1)
    .rolling(3)
    .mean()
)
```

```python
df["xG_last5"] = (
    df.groupby("player_id")["expected_goals"]
    .shift(1)
    .rolling(5)
    .mean()
)
```

```python
df["xA_last5"] = (
    df.groupby("player_id")["expected_assists"]
    .shift(1)
    .rolling(5)
    .mean()
)
```

```python
df = df.dropna()
```

```python
print(df[[
"player_id",
"round",
"total_points",
"avg_points_last3",
"avg_minutes_last3",
"xG_last5",
"xA_last5"
]].head(15))
```

```python
df = df.dropna().reset_index(drop=True)
```

```python
df["xGI_last5"] = (
    df.groupby("player_id")["expected_goal_involvements"]
    .shift(1)
    .rolling(5)
    .mean()
)
```

```python
df.to_csv("fpl_training_dataset.csv",index = False)
print("Dataset exported successfully")
print("Shape:", df.shape)
```
