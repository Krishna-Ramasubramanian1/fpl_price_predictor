##Average points from last 3


```python
df["avg_points_last3"] = (
    df.groupby("player_id")["total_points"]
    .shift(1)
    .rolling(3)
    .mean()
)
```

```python
df["avg_minutes_last3"] = (
    df.groupby("player_id")["minutes"]
    .shift(1)
    .rolling(3)
    .mean()
)
```

```python
df["xG_last5"] = (
    df.groupby("player_id")["expected_goals"]
    .shift(1)
    .rolling(5)
    .mean()
)
```

```python
df["xA_last5"] = (
    df.groupby("player_id")["expected_assists"]
    .shift(1)
    .rolling(5)
    .mean()
)
```

```python
df = df.dropna()
```

```python
print(df[[
"player_id",
"round",
"total_points",
"avg_points_last3",
"avg_minutes_last3",
"xG_last5",
"xA_last5"
]].head(15))
```

```python
df = df.dropna().reset_index(drop=True)
```

```python
df["xGI_last5"] = (
    df.groupby("player_id")["expected_goal_involvements"]
    .shift(1)
    .rolling(5)
    .mean()
)
```

```python
df.to_csv("fpl_training_dataset.csv",index = False)
print("Dataset exported successfully")
print("Shape:", df.shape)
```

## Player Performance Trends Over Time

### Subtask:
Select a few top-performing players (e.g., based on overall total points or high average points) and create line plots to show their 'total_points' and 'avg_points_last3' across different rounds. This will help identify trends and consistency.


**Reasoning**:
To select the top 5 players, I need to group the DataFrame by 'player_id', calculate the sum of 'total_points' for each player, and then identify the players with the highest total points.



```python
top_5_players = df.groupby('player_id')['total_points'].sum().nlargest(5).index.tolist()
print("Top 5 Player IDs based on total_points:", top_5_players)
```

**Reasoning**:
The next step is to filter the original DataFrame `df` to include only the data for the identified top 5 players. This will allow for easier plotting of their performance trends.



```python
df_top_players = df[df['player_id'].isin(top_5_players)].copy()
```

**Reasoning**:
Now I will iterate through the top 5 players and generate a line plot for each, showing their 'total_points' and 'avg_points_last3' over the rounds. This directly addresses the third instruction of the subtask.



```python
for player_id in top_5_players:
    player_data = df_top_players[df_top_players['player_id'] == player_id]

    plt.figure(figsize=(10, 6))
    sns.lineplot(data=player_data, x='round', y='total_points', label='Total Points', marker='o')
    sns.lineplot(data=player_data, x='round', y='avg_points_last3', label='Avg Points Last 3 Games', marker='x')
    plt.title(f'Player {player_id}: Total Points vs. Average Points Last 3 Games Over Rounds')
    plt.xlabel('Round')
    plt.ylabel('Points')
    plt.legend()
    plt.grid(True)
    plt.show()
```

## Further Correlation Analysis

### Subtask:
Expand the correlation analysis to include other important raw features such as 'goals_scored', 'assists', 'clean_sheets', and 'saves' with 'total_points', using a heatmap to visualize the relationships.


**Reasoning**:
